##### 1) Gender

##### 3) Marital_status

```python
ax =sns.countplot(x ="Marital_Status",data =df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ms =df.groupby(['Marital_Status','Gender'],as_index =False)['Amount'].sum().sort_values(by ='Amount',ascending= False)
sns.barplot(x='Marital_Status',y ='Amount',data=sales_ms ,hue ="Gender")
```

```python
From the above graph we can clearly see that most of the buyers are married (female)
```

##### 4) State

```python
sns.set(rc ={'figure.figsize':(10,5)})
ax =sns.countplot(x ="State",data =df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#From the above graph we can clearly see that most of the buyers from Uttarpradesh, Maharastra,Karnataka
```

```python
sales_state =df.groupby(['State'],as_index =False)['Amount'].sum().sort_values(by ='Amount',ascending= False).head(10)
sns.barplot(x='State',y ='Amount',data=sales_state)
```

```python
#From the above graph we can clearly see that most of the amount spent from Uttarpradesh, Maharastra,Karnataka
```

```python
sales_state =df.groupby(['State'],as_index =False)['Orders'].sum().sort_values(by ='Orders',ascending= False).head(10)
sns.barplot(x='State',y ='Orders',data=sales_state)
```

```python
#From the above graph we can clearly see that most of the orders from Uttarpradesh, Maharastra,Karnataka
```

##### 5) Occupations

```python
sns.set(rc ={'figure.figsize':(10,5)})
ax =sns.countplot(x ="Occupation",data =df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_oc =df.groupby(['Occupation'],as_index =False)['Orders'].sum().sort_values(by ='Orders',ascending= False).head(10)
sns.barplot(x='Occupation',y ='Orders',data=sales_oc)
```

```python
# From the above graph we can clearly see that most of the buyers are working in IT sector,Healthcare, Aviation
```

###### 6) Product_Category

```python
sns.set(rc ={'figure.figsize':(10,5)})
ax =sns.countplot(x ="Product_Category",data =df)
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
sales_pc =df.groupby(['Product_Category'],as_index =False)['Amount'].sum().sort_values(by ='Amount',ascending= False).head(10)
sns.barplot(x='Product_Category',y ='Amount',data=sales_pc)
```

```python
# From the above graph we can clearly see that most sold product from Food, clothing and then Electronics
```

```python
sales_pc =df.groupby(['Product_ID'],as_index =False)['Amount'].sum().sort_values(by ='Amount',ascending= False).head(10)
sns.barplot(x='Product_ID',y ='Amount',data=sales_pc)
```
